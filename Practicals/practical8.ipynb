```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# Load Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df['class'] = iris.target_names[iris.target]
```

```python
# a. Bar chart for frequency of each class label
plt.figure(figsize=(6, 4))
sns.countplot(x='class', data=iris_df)
plt.xlabel('Class Label')
plt.ylabel('Frequency')
plt.title('Frequency of Each Class Label')
plt.show()
```

```python
# b. Scatter plot for Petal width vs Sepal width
plt.figure(figsize=(6, 4))
sns.scatterplot(x='sepal width (cm)', y='petal width (cm)', data=iris_df)
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Width vs Sepal Width')
plt.show()
```

```python
# c. Density distribution for feature Petal Length
plt.figure(figsize=(6, 4))
sns.kdeplot(data=iris_df['petal length (cm)'], fill=True)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Density')
plt.title('Density Distribution of Petal Length')
plt.show()
```

```python
# d. Pair plot for pairwise bivariate distribution
sns.pairplot(iris_df, hue='class')
plt.suptitle('Pairwise Bivariate Distribution in Iris Dataset', y=1.02)
plt.show()
```
