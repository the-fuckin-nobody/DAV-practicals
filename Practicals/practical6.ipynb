```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {'Name': ['Mudit Chauhan', 'Seema Chopra', 'Rani Gupta', 'Aditya Narayan', 'Sanjeev Sahni', 'Prakash Kumar', 'Ritu Agarwal', 'Akshay Goel', 'Meeta Kulkarni', 'Preeti Ahuja', 'Sunil Das Gupta', 'Sonali Sapre', 'Rashmi Talwar'],
        'Birth_Month': ['December', 'January', 'March', 'October', 'February', 'December', 'September', 'August', 'July', 'November', 'April', 'January', 'June'],
        'Gender': ['M', 'F', 'F', 'M', 'M', 'M', 'F', 'M', 'F', 'F', 'M', 'F', 'F'],
        'Pass_Division': ['III', 'II', 'I', 'I', 'II', 'III', 'I', 'I', 'II', 'II', 'III', 'I', 'III']}

df = pd.DataFrame(data)
df
```

```python
class_counts = df['Pass_Division'].value_counts()
class_counts
```

```python
class_counts.plot(kind='bar')
plt.show()
```

```python
df['Pass_Division'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
df['Birth_Month_Num'] = pd.to_datetime(df['Birth_Month'], format='%B').dt.month
sns.kdeplot(df['Birth_Month_Num'])
plt.show()
```
